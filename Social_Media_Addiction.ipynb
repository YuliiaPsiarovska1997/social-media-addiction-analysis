```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Addicted_Score'], kde=True)
plt.title("Distribution of Addicted Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title("Distribution of Students' Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Gender', hue='Gender', data=df, palette='Set2', legend=False)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.boxplot(x='Affects_Academic_Performance', y='Sleep_Hours_Per_Night', data=df)
plt.title("Effect of Sleep Duration on Academic Performance")
plt.xlabel("Impact on Academic Performance")
plt.ylabel("Hours of Sleep per Night")
plt.show()
```

```python
sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True, color='orange')
plt.title("Average Daily Time Spent on Social Media")
plt.xlabel("Hours per Day")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Most_Used_Platform', hue='Most_Used_Platform', data=df, palette='viridis', legend=False)
plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Mental_Health_Score', data=df, color='purple')
plt.title("Relationship between Time on Social Media and Mental Health")
plt.xlabel("Hours per Day on Social Media")
plt.ylabel("Mental Health Score")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Addicted_Score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlations Between Numerical Indicators")
plt.show()
```

```python
sns.countplot(x='Conflicts_Over_Social_Media', hue='Relationship_Status', data=df, palette='muted')
plt.title("Conflicts Due to Social Media and Relationship Status")
plt.xlabel("Conflicts")
plt.ylabel("Number of Students")
plt.show()
```

### Final Conclusion â€“ Social Media Usage & Student Well-being

This analysis explores social media usage patterns among 705 students, examining its impact on sleep, academic performance, and mental health.

---

#### Social Media Usage
- Average daily usage is approximately 4.92 hours.
- The most frequently used platform is Instagram.
- Undergraduate students show the highest usage.

---

#### Mental Health & Addiction
- Students with higher Addicted_Score:
  - Sleep less (especially those with <6 hours per night).
  - Report lower mental health scores.
  - Experience more interpersonal conflicts related to social media.

---

#### Academic Impact
- Students who report that social media negatively affects academic performance:
  - Use it more often.
  - Have lower sleep duration.
  - Tend to be more addicted (higher Addicted_Score).

---

#### Social Conflicts
- Conflicts over social media are more common among younger students.
- Students with an Addicted_Score above 7 frequently experience more social friction.

---

### Implications for Institutions
- Excessive social media use is associated with measurable negative effects on sleep, well-being, and academic outcomes.
- Educational institutions should consider:
  - Implementing awareness programs.
  - Providing mental health and digital wellness support.
  - Including behavioral data monitoring and education on responsible technology use.

---

*These findings highlight the importance of understanding digital behavior and its broader impact on student life.*
